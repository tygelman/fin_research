```python
import yfinance as yf
df = yf.download(tickers="MSFT",
                            period="1d",
                            interval="1m",
                            auto_adjust=True)
df.head(10)
```

```python
df["MA5"] = df['Close'].rolling(5).mean()
df["MA20"] = df['Close'].rolling(20).mean()
#df['EMA'] = df['Close'].ewm(halflife=5, min_periods=20).mean()
```

```python
import matplotlib.pyplot as plt
#df['Close'].plot(label = 'MSFT_Close', figsize = (16, 8), title = 'Adjusted 5m Price')
#df['Open'].plot(label = 'MSFT_Open', figsize = (16, 8), title = 'Adjusted 5m Price')

df[['Close', 'MA5', 'MA20']].plot(label = 'MSFT', figsize = (16, 8))
plt.legend()
```
